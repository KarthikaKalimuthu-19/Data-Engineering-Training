Read it into a pandas DataFrame and print the first 3 rows.

```python
import pandas as pd

df = pd.read_csv(r'C:\Users\ELCOT\Desktop\students.csv')
print(df.head(3))
```

2. Calculate Average Scores
Add a new column Average that calculates the mean score for Math, Science, and English for each student.

    Print the student with the highest average.

```python
df['Average'] = df[['Math', 'Science', 'English']].mean(axis=1)
print("\nStudent with the highest average:")
print(df[df['Average'] == df['Average'].max()])
```

3. Filter Students
Display all students who have scored more than 80 in Science.
Display all students with an average score below 70.

```python
mark = df[df['Science'] > 80]
print("Students scored more than 80 in Science: \n",mark)

mark1 = df[df['Average'] < 70]
print("Students with an average score below 70: \n",mark1)
```

4. Sorting
Sort the students by English score (descending) and print the top 3.

```python
sort_by_eng = df.sort_values(by='English', ascending=False).head(3)
print("the students by English score top 3: \n",sort_by_eng)
```

5. Chart Exercise
Create a bar chart showing students’ names on the X-axis and their average
scores on the Y-axis.
Create a line chart for all students’ scores in Math, Science, and English (use
df.plot() ).

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(df['Name'], df['Average'], color='pink')
plt.title('Student Average Scores')
plt.xlabel('Name')
plt.ylabel('Average Score')
plt.grid(axis='y')
plt.show()
```

```python
df.set_index('Name')[['Math', 'Science', 'English']].plot(marker='o')
plt.title('Subject-wise Scores')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```

6. Bonus Task
Find the subject with the highest total score across all students.
Save the updated DataFrame (with Average column) as a new CSV named
students_summary.csv .

```python
tot = df[['Math', 'Science', 'English']].sum()
top = tot.idxmax()

print("\nSubject with the highest total score:", top)

df.to_csv(r'C:\Users\ELCOT\Desktop\students_summary.csv', index=False)
print("\nSaved updated DataFrame as 'students_summary.csv'")
```
